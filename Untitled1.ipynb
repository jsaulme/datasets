```python
import pandas as pd
```

```python
moves = pd.read_csv("https://raw.githubusercontent.com/jsaulme/datasets/main/moves_df.csv")
moves.head()
```

```python
# Agrega una columna llamada "turns" y que sea igual a "cooldown"/500

moves["turns"] = moves["cooldown"]/500
moves.head()
```

```python
# Aplica formato "int" a la columna "turns"
moves["turns"] = moves["turns"].astype(int)
moves.head()
```

```python
# Agrega una columna llamada "energy_per_turn" y que sea igual a "energy"/"turns" en decimales
moves["energy_per_turn"] = moves["energyGain"]/moves["turns"]
moves.head()
```

```python
# Agregar columna "damage_per_energy" que sea igual a "power" / "energy" y que contenga solo 2 decimales
moves["damage_per_energy"] = moves["power"]/moves["energy"]
moves["damage_per_energy"] = moves["damage_per_energy"].round(2)
moves.head()
```
